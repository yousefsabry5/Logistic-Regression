```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
columns_to_encode = ['Name', 'Sex', 'Ticket', 'Cabin', 'Embarked']

for column in columns_to_encode:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
from sklearn.preprocessing import StandardScaler

# Select features
features = df.drop('Survived', axis=1)
target = df['Survived']

# Normalize the df
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Convert to dfFrame
features_scaled = pd.DataFrame(features_scaled, columns=features.columns)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Initialize and train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Accuracy of the model: {:.2f}%".format(accuracy * 100))
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

```python
import numpy as np
import pandas as pd
X = df[['Pclass', 'Age', 'Sex']]
y = df['Survived']


```

```python
X = X.values
y = y.values
```

```python
train_ratio = 0.8
test_ratio = 1 - train_ratio

num_train = int(train_ratio * len(X))
num_test = len(X) - num_train

X_train = X[:num_train]
X_test = X[num_train:]
y_train = y[:num_train]
y_test = y[num_train:]
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def hypothesis(X, theta):
    z = np.dot(X, theta)
    return sigmoid(z)
```

```python
def cost_function(X, y, theta):
    m = len(y)
    h = hypothesis(X, theta)
    cost = (-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    costs = []
    for i in range(iterations):
        h = hypothesis(X, theta)
        gradient = np.dot(X.T, (h - y)) / m
        theta -= alpha * gradient
        cost = cost_function(X, y, theta)
        costs.append(cost)
    return theta, costs

theta = np.zeros(X.shape[1])
alpha = 0.01
iterations = 1000

theta, costs = gradient_descent(X, y, theta, alpha, iterations)
```

```python
def accuracy_score(y_true, y_pred):
    correct = np.sum(y_true == y_pred)
    total = len(y_true)
    accuracy = correct / total
    return accuracy
```

```python
y_pred = hypothesis(X_test, theta)
y_pred_class = (y_pred >= 0.5).astype(int)
```

```python
def confusion_matrix(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))  # True Positives
    TN = np.sum((y_true == 0) & (y_pred == 0))  # True Negatives
    FP = np.sum((y_true == 0) & (y_pred == 1))  # False Positives
    FN = np.sum((y_true == 1) & (y_pred == 0))  # False Negatives

    return np.array([[TN, FP], [FN, TP]])

```

```python

cm = confusion_matrix(y_test, y_pred_class)
print('Confusion Matrix:')
print(cm)
```

```python
precision = cm[1, 1] / (cm[1, 1] + cm[0, 1])

recall = cm[1, 1] / (cm[1, 1] + cm[1, 0])

print(f' (Precision): {precision}')
print(f' (Recall): {recall}')
```
